## Description of the data set

The dataset you mentioned is a state-wise record of COVID-19 vaccinations in India. It contains information on the number of doses administered, the number of sessions and sites, the number of people who received the first and second doses, and the number of people vaccinated by gender and vaccine type. It also includes information on adverse events following immunization (AEFI) and the age distribution of those vaccinated.

```python
latest_data = df.groupby('State').last().reset_index()
```

```python
# Number of persons state wise vaccinated for first dose in India
first_dose = latest_data[['State', 'First Dose Administered']]
latest_data = latest_data[latest_data['State'] != 'India']
print(first_dose)
```

```python
# Number of persons state wise vaccinated for second dose in India
latest_data = latest_data[latest_data['State'] != 'India']
second_dose = latest_data[['State', 'Second Dose Administered']]
print(second_dose)
```

```python
plt.bar(second_dose['State'], second_dose['Second Dose Administered'])
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Second Dose Administered')
plt.show()
```

```python
# Number of Males vaccinated
male_vaccinated = latest_data['Male (Doses Administered)'].sum()
print(male_vaccinated)
```
